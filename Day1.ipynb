```python
#WAP to data type and address of variables
math=40
pi=3.14
name="shweta sm"
result=True
print(type(math)) #type function is used for data type
print(type(pi))
print(type(name))
print(type(result))
print(id(math)) # id function is for address of variable
print(id(pi))
print(id(name))
print(id(result))
```

```python
#WAP to swap to numbers using six bytes(using 3 variables)
val1=20
val2=40
print("Before swapping ",val1," ",val2)
temp=val1
val1=val2
val2=temp
print("After swapping ",val1," ",val2)
```

```python
#WAP to swap to numbers using four bytes(using 2 variables)
val1=20
val2=40
print("Before swapping ",val1," ",val2)
val1=val1+val2
val2=val1-val2
val1=val1-val2
print("After swapping ",val1," ",val2)
```
