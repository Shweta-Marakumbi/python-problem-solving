```python
#wap to find maximum and minimum elements in array
a=[5,3,9,2,8,10]
max=a[0]
min=a[0]
for i in a:
    if i>max:
        max=i
    if i<min:
        min=i
print("Maximum = ",max)
print("Minimum = ",min)
```
