```python
#wap to present character in string. it will write index and character 
name="shweta"
i=0
for x in name:
    if x=='t':
        print("The character present at index no",i,"value=:",x)
        break
    i=i+1
print(i)
```

```python
#wap to count vowels and consonants
name="shweta"
vow=0
cons=0
vowels=['a','e','i','o','u']
for i in name:
    if i in vowels:
        vow +=1
    else:
        cons+=1
print("vowels=",vow)
print("consonants=",cons)
```
