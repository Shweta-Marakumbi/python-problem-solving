```python
#chr() function typecast integer value in character
n=int(input("Enter the number of rows:  ")) #5
for i in range(1,n+1): #i = row
    for j in range(1,n+1): #j=colm
        print(chr(64+i),end="  ") #ASCII 
    print()
```

```python
#positional argument # in python is possible to create multiple values at a atime
#Addition of two numbers using calling function
def add(x,y):
    add=x+y
    sub=x-y
    mul=x*y
    div=x/y
    return add,sub,mul,div # return in tuple
a=int(input("Enter the value of A: "))
b=int(input("Enter the value of B: "))
result=add(a,b)
print("Ans= ",result)
```

```python
#handling multiple different kinds of exception with single except block
try:
    a=int(input("Enter any one number"))
    b=int(input("Enter any second number"))
    print(a/b)
except (ValueError,ZeroDivisionError)as message:
    print(message)
```

```python
#we can use multiple except block
try:
    a=int(input("Enter any one number"))
    b=int(input("Enter any second number"))
    print(a/b)
except(ValueError,ZeroDivisionError)as message:
    print("Enter correct number:",message)
except:
    print("This is default part of except block")
```

```python
#we can use else block if no error will generate it is depend on our own need and necessity
try:
    a=int(input("Enter any one number"))
    b=int(input("Enter any second number"))
    print(a/b)
except(ValueError,ZeroDivisionError)as message:
    print("Enter correct number:",message)
else:
    print("Everything is ok")
```

```python
#Finally blocks will always executed whwther try block generate error or not
try:
    a=int(input("Enter any one number"))
    b=int(input("Enter any second number"))
    print(a/b)
except(ValueError,ZeroDivisionError)as message:
    print("Enter correct number:",message)
finally:
    print("I will always executed")
```

```python
#nested try except block
try:
    a=int(input("Enter any one number"))
    b=int(input("Enter any second number"))
    try:
        print(a/b)
    except ZeroDivisionError as msg:
        print(msg)
except ValueError as mes:
    print(msg)
```

```python
#exception handling
try:
    a=int(input("Enter any one number"))
    b=int(input("Enter any second number"))
    print(a/b)
except(ValueError,ZeroDivisionError)as message:
    print(message)
else:
    print("There are no error in try block")
finally:
    print("I am finally block I always executed whether error raise or not")
```
