```python
#naming convention rule=classname start with capital letter
class Student:
    #data member
    a=10
    b=20
    #self is default argument
    def add(self):
        print(self.a+self.b)
#object creation outside the class
obj=Student()
obj.add() 
print(obj.a)#calling method we use "."
print(obj.b)
#by default everything in class like variable and function are public
```

```python
#default constructor
#constructor is to calculate how much memory is required to create object.constructor call authomatically.
#internaly we have default constructor .one obj it called only one time.
class NewClass:
    def __init__(self):  # constructor (called automatically)
        print("constructor always executes first")

    def show(self):  # member function inside class
        print("Welcome to class level programming")

obj = NewClass()  # creating an object
obj.show()

obj1 = NewClass()  # creating another object
obj1.show()
    
```

```python
#how many types of variables we declare in class
#static variable #local variable
#we can declare instance variable using self keyword in constructor
class Student:
    def __init__(self):
        self.s_name="shweta" #instance variable
        self.s_age=21
        self.s_branch="CS"
    def getdata(self):  #instance method
        self.s_mb=28282828282  #instance variable
obj=Student()
obj.getdata()
obj.s_branch="CS" #outside class instace variable by using object
print(obj.__dict__)
print(obj.s_mb)
        
```

```python
#for every object instace varible create separate memory
class New:
    def __init__(self):
        self.a=10 #instance variable
Obj1=New() # each obj create separate memory
Obj2=New()
Obj3=New()
Obj1.a=20 # only this obj1 value will change bcz 20 is storing in obj1 
print(Obj1.a)
print(Obj2.a)
print(Obj3.a)
```

```python
# one static varible=one memory it is exactly reverse of instance variable
#there are 2 ways to access static variable the first one is by class name or second way by object 
#by using object or self we can access static variable
#static varible does not depend on object
class New:
    a=10 #static variable
    def __init__(self):
        self.name="shweta"
Obj1=New()
Obj2=New()
Obj3=New()
print(Obj1.a)
print(Obj2.a)
print(Obj3.a)
New.a=20     #using class name we are changing value
print(Obj1.a)
print(Obj2.a)
print(Obj3.a)
```

```python
class Student:
    def __init__(self,name,rollno,mob):
        self.name=name  #instance variable
        self.rollno=rollno
        self.mob=mob
    def display(self):  #instance method
        print(self.name,"",self.rollno,"",self.mob)
stud=Student("shweta",33,1234567897)
stud.display()
```

```python
#static method-----------------------------------
class Student:
    #by using class name we can access static method
    @staticmethod #decorate
    def get_personal_detail(firstname,lastname):
        print("your personal details=",firstname,lastname)
    @staticmethod
    def contact_detail(mobil_no,rollno):
        print("your contact detail=",mobil_no,rollno)
Student.get_personal_detail("shweta","marakumbi")
Student.contact_detail(1234567898,1001)
```

```python
# #Inheritance---------
# Extending property from one another class is called inheritance.
# Directly we are getting here reuseability concept
# 1 Base class: A class which inherit its property to another is called base class or parent class
# 2 Derived cless: A class in which properties are inherit called as derived class or child class
# Types of inheritance:
# 1 Single Inheritance
# 2 Multilevel Inheritance
# 3 Multip[le Inheritance
# for single inheritance
# syntax:   class derived-class(base-class):
#                  ------------------------
#                  ------------------------
```

```python
#singlelevel Inheritance-----
class College:
    def college_name(self): #member function of college
        print("Modern college")
class Student(College):#child class
    def student_info(self):  #member function
        print("Name:  shweta")
        print("Branch:  CSE")
obj=Student()#object create child class
obj.college_name()
obj.student_info()
```

```python
# #Multilevel Inheritance----------
# syntax: class GrandFather:
#               --------------
#               --------------
#         class Father(GrandFather):
#             --------------------
#             --------------------
#         class Child(Father):
#             ------------------
#             ------------------
class Collage:  #first class first-level
    def college_name(self):
        print("Modern College")
#--------------------------------------------
class Student(College):#second class second-level
    def student_info(self):
        print("Name: shweta marakumbi")
        print("Branch: CSE")
#-------------------------------------------------
class Exam(Student):#child class
    def subject(self):
        print("subject1: Design Engineering")
        print("subject2: Math")
        print("subject3: c-language")
obj=Exam()
obj.college_name()
obj.student_info()
obj.subject()
```

```python
#Multiple Inheritance-----------------------
#multiple inheritance means a class can inherit from more than one parent class at the same time.
class SubjMarks:#class-1
    math = int(input("Enter paper marks of math:"))
    DE = int(input("Enter paper marks of Design Engineering"))
    C = int(input("Enter Paper marks of c language"))
    english = int(input("Enter oaoer marks of english"))
#-------------------------------------------------------parent class-1
class PractMarks:#class-
    cpract = int(input("Enter practical marks of c language"))
#---------------------------------------------------------parent class-2
class Result(SubjMarks,PractMarks):#child class
    #print("if student pass in both = subject and practicle paper then pass")
    def total(self):
        if self.math>=40 and self.DE>=40 and self.C>=40 and self.english>=40 and self.cpract>=20:
            print("pass")
        else:
            print("fail")
obj = Result()
obj.total()
```

```python
#Polymorphism------------------------
# Polymorphism means having many forms.
# We can define polymorphism as the ability of a message to be displayed in more than one form.
# In C++, there are two types of polymorphism:
# Compile time polymorphism (e.g., Overloading)
# Runtime polymorphism (e.g., Overriding)
# polymorphism example
class Principal:
    def role(self):
        print("I am managing all activity of college")
class Dean:
    def role(self):
        print('Dean = I am decision taking person')
class Hod:
    def role(self):
        print('Hod = I have responsibility of Teachers and Students')
class Faculty:
    def role(self):   
        print('Faculty = I have to complete syllabus successfully')
# ========== class declaration completed ==============================
def func(obj):
    obj.role()   # calling same method for all objects
campus = [Principal(), Dean(), Hod(), Faculty()]
for obj in campus:
    func(obj)
```
